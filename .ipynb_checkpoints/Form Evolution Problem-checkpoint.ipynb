```python
import numpy as np

#Plotting Scatter 3d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Fast Sum of Gaussians Kernels
from pyfigtree import figtree

#Using For ODE
from scipy.optimize import fsolve
from scipy.integrate import odeint

%matplotlib inline

epsilon = 1e-5
```

```python
def direct_evaluation(S,T,weights,sigma):
    checkPot = []
    for target in T:
        # for each source, add to potential at current target
        diff = target-S
        norm = np.linalg.norm(diff, axis=1)
        up = (norm / sigma)**2
        potential = np.exp(-up)*weights
        
        cum_potential = np.sum(potential)
        checkPot.append(cum_potential)
    return np.array(checkPot)
```

# Form Evolution Equations
## Direct Functions

```python
def direct_position(X,alpha,sigma):
    X_x = direct_evaluation(X,X,alpha[:,0], sigma)
    X_y = direct_evaluation(X,X,alpha[:,1], sigma)
    X_z = direct_evaluation(X,X,alpha[:,2], sigma) 
    return np.dstack([X_x,X_y, X_z])[0]
```

```python
def direct_momentum(X,alpha,sigma):
    ax = alpha[:,0]
    ay = alpha[:,1]
    az = alpha[:,2]
    
    #alpha_xj * X_j
    first_weights = np.dstack([ax * X[:,0], ax * X[:,1], ax * X[:,2]])[0]
    
    #alpha_yj * X_j
    second_weights = np.dstack([ay * X[:,0], ay * X[:,1], ay * X[:,2]])[0]
    
    #alpha_zj * X_j
    third_weights = np.dstack([az * X[:,0], az * X[:,1], az * X[:,2]])[0]
    
    #alpha_xi * sum( alpha_xj * X_j )
    first_sum_x = ax* direct_evaluation(X,X,first_weights[:,0],sigma)
    first_sum_y = ax* direct_evaluation(X,X,first_weights[:,1],sigma)
    first_sum_z = ax* direct_evaluation(X,X,first_weights[:,2],sigma)
    
    #alpha_yi * sum( alpha_yj * X_j )    
    second_sum_x = ay * direct_evaluation(X,X,second_weights[:,0],sigma)
    second_sum_y = ay * direct_evaluation(X,X,second_weights[:,1],sigma)
    second_sum_z = ay * direct_evaluation(X,X,second_weights[:,2],sigma)
    
    #alpha_zi * sum( alpha_zj * X_j )        
    third_sum_x = az * direct_evaluation(X,X,third_weights[:,0],sigma)
    third_sum_y = az * direct_evaluation(X,X,third_weights[:,1],sigma)
    third_sum_z = az * direct_evaluation(X,X,third_weights[:,2],sigma)
    
    #sum(alpha_x)
    fourth_sum = direct_evaluation(X,X,ax,sigma)
    fifth_sum = direct_evaluation(X,X,ay,sigma)
    sixth_sum = direct_evaluation(X,X,az,sigma)
    
    second_term = ax * fourth_sum + ay * fifth_sum + az * sixth_sum

    
    mult = -2./(sigma ** 2)
    result_x = mult*(first_sum_x + second_sum_x + third_sum_x - (X[:,0] * second_term) )
    result_y = mult*(first_sum_y + second_sum_y + third_sum_y - (X[:,1] * second_term) )
    result_z = mult*(first_sum_z + second_sum_z + third_sum_z - (X[:,2] * second_term) )
    result = np.dstack([result_x,result_y,result_z])[0]
    
    return result
```

## Figtree Functions

```python
def position(X,alpha,sigma):
    X_x = figtree(X,X,alpha[:,0], sigma,epsilon=epsilon, eval="auto")
    X_y = figtree(X,X,alpha[:,1], sigma,epsilon=epsilon, eval="auto")
    X_z = figtree(X,X,alpha[:,2], sigma,epsilon=epsilon, eval="auto") 
    return np.dstack([X_x,X_y, X_z])[0]
```

```python
def momentum(X,alpha,sigma):
    ax = alpha[:,0]
    ay = alpha[:,1]
    az = alpha[:,2]
        
    first_weights = np.dstack([ax * X[:,0], ax * X[:,1], ax * X[:,2]])[0]
    second_weights = np.dstack([ay * X[:,0], ay * X[:,1], ay * X[:,2]])[0]
    third_weights = np.dstack([az * X[:,0], az * X[:,1], az * X[:,2]])[0]
    
    first_sum_x = ax * figtree(X,X,first_weights[:,0],sigma, epsilon=epsilon, eval="auto")
    first_sum_y = ax * figtree(X,X,first_weights[:,1],sigma, epsilon=epsilon, eval="auto")
    first_sum_z = ax * figtree(X,X,first_weights[:,2],sigma, epsilon=epsilon, eval="auto")
    
    second_sum_x = ay * figtree(X,X,second_weights[:,0],sigma, epsilon=epsilon, eval="auto")
    second_sum_y = ay * figtree(X,X,second_weights[:,1],sigma, epsilon=epsilon, eval="auto")
    second_sum_z = ay * figtree(X,X,second_weights[:,2],sigma, epsilon=epsilon, eval="auto")
    
    third_sum_x = az * figtree(X,X,third_weights[:,0],sigma, epsilon=epsilon, eval="auto")
    third_sum_y = az * figtree(X,X,third_weights[:,1],sigma, epsilon=epsilon, eval="auto")
    third_sum_z = az * figtree(X,X,third_weights[:,2],sigma, epsilon=epsilon, eval="auto")
    
    fourth_sum  = figtree(X,X,ax,sigma, epsilon=epsilon, eval="auto")
    fifth_sum   = figtree(X,X,ay,sigma, epsilon=epsilon, eval="auto")
    sixth_sum   = figtree(X,X,az,sigma, epsilon=epsilon, eval="auto")
    
    second_term = ax * fourth_sum + ay * fifth_sum + az * sixth_sum
    
    second_term_x = X[:,0] * second_term
    second_term_y = X[:,1] * second_term
    second_term_z = X[:,2] * second_term
 
    mult = -2./(sigma ** 2)
    result_x = mult*(first_sum_x + second_sum_x + third_sum_x - (second_term_x) )
    result_y = mult*(first_sum_y + second_sum_y + third_sum_y - (second_term_y) )
    result_z = mult*(first_sum_z + second_sum_z + third_sum_z - (second_term_z) )
    result = np.dstack([result_x,result_y,result_z])[0]
    
    return result
```

# Linearizing Data

```python
def f(vector, sigma):
    positions, momentums = np.split(vector,2)
    positions = np.reshape(positions,(3,len(positions)/3)).T
    momentums = np.reshape(momentums,(3,len(momentums)/3)).T
    new_positions = position(positions, momentums,sigma)
    new_momentums = momentum(positions, momentums,sigma)
    n,m = new_positions.shape
    new_positions = np.reshape(new_positions.T, (n*m))
    new_momentums = np.reshape(new_momentums.T, (n*m))
    return np.hstack([new_positions,new_momentums])
```

```python
def f_direct(vector, sigma):
    positions, momentums = np.split(vector,2)
    positions = np.reshape(positions,(3,len(positions)/3)).T
    momentums = np.reshape(momentums,(3,len(momentums)/3)).T
    new_positions = direct_position(positions, momentums,sigma)
    new_momentums = direct_momentum(positions, momentums,sigma)
    n,m = new_positions.shape
    new_positions = np.reshape(new_positions.T, (n*m))
    new_momentums = np.reshape(new_momentums.T, (n*m))
    return np.hstack([new_positions,new_momentums])
```

# Shooting Method

```python
def shoot(alpha, x_a, x_b, sigma,steps):
    #Making lambda function to integrate
    g = lambda y,t: f(y,sigma)
    
    x_i = x_a
    a_i = alpha
    
    #Linealizing vectors
    n,m = x_i.shape
    x_i = np.reshape(x_i.T, (n*m))
    #a_i = np.reshape(a_i.T, (n*m))
    
    vector = np.hstack([x_i,a_i])
    #Integrate over time
    t = np.linspace(0,1,steps)
    matrix = odeint(g,vector,t)
    
    vector = matrix[-1]
    x_b = np.reshape(x_b.T, (n*m))
    x_i, momentums = np.split(vector,2)
    return x_i - x_b
```

```python
def shootingMethod(x_a, x_b, alpha0, steps, sigma):
    alpha = fsolve(shoot, alpha0, (x_a,x_b,sigma,steps))
    return alpha
```
